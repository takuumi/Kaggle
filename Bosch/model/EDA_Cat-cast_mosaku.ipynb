```python
CHUNK_SIZE = 10000
start = datetime.now()

reader_temp = pd.read_csv("train_categorical.csv",chunksize=CHUNK_SIZE, index_col=0, dtype='object')

#分割して読み込み
loop = 0
for df_chunk_res in reader_temp:
    df_chunk_cat = reader_temp.get_chunk(CHUNK_SIZE)
    
    for i in range(0, len(df_chunk_cat.columns)):
        df_chunk_cat.iloc[:,i] = df_chunk_cat.iloc[:,i].str[1:].astype(float).fillna(0).astype(int)
                    
    if loop == 0:
        df_tst = df_chunk_cat
    else:
        df_tst = pd.concat([df_tst, df_chunk_cat])

    loop = loop + 1

                   
    print("loop : {} / Rows = {}".format(loop, len(df_tst.index)))
    print("{} -> {}".format(start, datetime.now()))
    
    if (loop == 2):
        break
```
